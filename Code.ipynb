```python
import pandas as pd

# قراءة الملف (تأكد الاسم صحيح)
df = pd.read_csv(r"C:\Users\AL-Am\Downloads\archive\superstore.csv")


# عرض البيانات
print(df.head())

# حذف القيم الفارغة
df = df.dropna()

# تجميع حسب المنتج
top_products = df.groupby("Product.Name")["Sales"].sum().sort_values(ascending=False)

print(top_products.head())


# أفضل العملاء
top_customers = df.groupby("Customer.Name")["Sales"].sum().sort_values(ascending=False)

print(top_customers.head())
```
